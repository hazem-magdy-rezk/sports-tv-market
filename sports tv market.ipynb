## the last step is the analysis of the count of teams participating in all four leagues for each market :

```python
# grouping the data based on total number of teams participating in each leaby homes and markets
# total number of teams is not a sum its based on considering 1 team for each league
df['total_teams'] = df[['NBA','NFL','MLB','NHL']].gt(0).sum(axis=1)
grouping_by_homes=df.groupby(['total_teams'])['Homes'].sum()
grouping_by_Market=df.groupby(['total_teams'])['Market'].count()
```

```python
# dataFraming the grouped data
grouped_df=pd.DataFrame()
grouped_df['total_teams']=np.sort(df['total_teams'].unique())
grouped_df['count_of_markets']=np.array(grouping_by_Market)
grouped_df['sum_of_homes']=np.array(grouping_by_homes)
```

```python
# print the DataFrame
grouped_df
```

```python
# plotting the grouped data 
labels=grouped_df['total_teams']
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, grouped_df['sum_of_homes'], width, label='homes')
rects2 = ax.bar(x + width/2, grouped_df['count_of_markets'], width, label='markets')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('number of homes')
ax.set_title('number of teams')
ax.bar_label(rects1, padding=3,fmt="%i")
ax.bar_label(rects2, padding=3,fmt="%i")
ax.set_xticks(labels)
ax.ticklabel_format(style='plain')
ax.ticklabel_format(useOffset=False)
fig.tight_layout()
plt.show()

```

* 14 markets have 1 team ( 14.7M homes )
* 8 markets have 2 teams ( 7.8M homes )
* 7 markets have 3 teams ( 13 M homes)
* These are 29 markets with count of 35.5M homes
* The proportion of these 35.5M home is 0.47 meaning almost the half 
* 35M household don’t have at least 1 team for each league


```python
# ploting the total number of households being covered/uncovered by tv-markets
plt.pie(x=[grouped_df[grouped_df['total_teams']==4]['sum_of_homes'].sum(),\
grouped_df[grouped_df['total_teams']<4]['sum_of_homes'].sum()],
        labels=[['covered ',grouped_df[grouped_df['total_teams']==4]['sum_of_homes'].sum()],
                ['uncovered',grouped_df[grouped_df['total_teams']==4]['sum_of_homes'].sum()]],
                autopct='%1.1f%%',radius=2);
plt.show()
```

35M household don’t have at least 1 team for each league
